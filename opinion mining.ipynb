```python
import nltk
import pandas as pd
```

```python
#opening the file
data=open('C:/Users/nagap/Desktop/stplist.txt',encoding="utf8")
```

```python
#reading the file
reviewsList=data.read()
```

```python
#removing the unnecessary parts
cleaneddata=''
for words in reviewsList:
    if words not in ['\n','"']:
        cleaneddata=cleaneddata+words
```

```python
#word Tokenizing
wordTokenized=nltk.word_tokenize(cleaneddata)
wordTokenized
```

```python
#sentence tokenizing
sentTokenized=nltk.sent_tokenize(cleaneddata)
sentTokenized
```

```python
x=nltk.pos_tag(sentTokenized) #not clear
```

```python
#framing the pos tags for each and every sentence
PatternFrame=[]
for sent in sentTokenized:
    wordbreak=nltk.word_tokenize(sent)
    PatternFrame.append(nltk.pos_tag(wordbreak))
PatternFrame
```

```python
#linguistic patterns generation
patterns=[]
for textTup in PatternFrame:
    string1=''
    for unitpattern in textTup:
        if unitpattern[1]!='.':
            string1=string1+unitpattern[1]+' '
    patterns.append(string1)
        
patterns       
```

```python
from collections import Counter
```

```python
#frequenctly occured patterns
allPatterns=Counter(patterns).most_common(15)
```

```python
#Taking top 15 patterns for comparision
PatternRules=[]
for i in allPatterns:
    PatternRules.append(i[0])
```

```python
PatternRules
```

```python
#creating dictionary mapping the sentence and the pattern for the sentence
sent_with_patterns=dict(zip(sentTokenized,patterns))
```

```python
#extracting the important sentences which are matching the top 15 linguistic patterns
Final_imp_sentences=[]
for i,j in sent_with_patterns.items():
    if j in PatternRules:
        Final_imp_sentences.append(i)
```

```python
#printing the matching sentences w.r.t to rule
Final_imp_sentences
```

```python
#extracting the aspects from the matched sentences
Aspects=[]
for i in Final_imp_sentences:
    words=nltk.word_tokenize(i)
    for j in nltk.pos_tag(words):
        if j[1]=='NN' or j[1]=='NNS' or j[1]=='NNP' or j[1]=='NNPS':
            Aspects.append(j[0])
            
Aspects=list(set(Aspects))
Aspects
```

```python
#all aspects from the comments
DummyAspects=[]
for (j) in nltk.pos_tag(wordTokenized):
    if j[1]=='NN' or j[1]=='NNS' or j[1]=='NNP' or j[1]=='NNPS':
        DummyAspects.append(j[0])
print(set(DummyAspects))
```

```python
#mapping the real aspects from the generated aspects and taking out the real matched aspects
realFeature=['echo','setup','screen','display','bluetooth','music','camera','video','cable','speaker','price','button','switch','power','wifi','battery','processor','order','memory','microphone','games','show','Google','message']
for i in Aspects:
    if i in realFeature:
        print(i)
```
